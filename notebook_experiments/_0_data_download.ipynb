```python
import pandas as pd
```

```python

roster_files = [
    f'https://github.com/nflverse/nflverse-data/releases/download/rosters/roster_{yr}.csv'
    for yr in range(2011,2025) 
]
```

```python
all_data = pd.DataFrame()
for file in roster_files:
    print(file)
    data = pd.read_csv(file)
    all_data = pd.concat([all_data,data])
```

```python
all_data.to_csv("../data/rosters.csv",index=False)
```

```python
stats_files = [
    f'https://github.com/nflverse/nflverse-data/releases/download/player_stats/player_stats_{yr}.csv'
    for yr in range(2010,2024) 
]
```

```python
all_data = pd.DataFrame()
for file in stats_files:
    print(file)
    data = pd.read_csv(file)
    all_data = pd.concat([all_data,data])

all_data.to_csv("../data/stats.csv",index=False)
```
